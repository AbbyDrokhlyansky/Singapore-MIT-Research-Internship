```python
#use this
import pandas as pd
from glob import glob
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#use this
df=pd.DataFrame()
filenames = glob("DataGovSG/*_rainfall.csv")
for filename in filenames:
    other_df = pd.read_csv(filename)
    other_df = other_df.rename(columns={"value": "rainfall"})
    other_df["timestamp_sgt"] = other_df["timestamp_sgt"].astype("datetime64[ns]")
    df = df.append(other_df,ignore_index=True)
another_df = df.set_index("station_id").sort_index()
```

```python
#use this
timechange = pd.to_timedelta("01:00:00")
```

```python
#use this
nnon_zero = another_df.sort_values("timestamp_sgt",ascending=True).reset_index().set_index("timestamp_sgt",drop=False)
times = nnon_zero.index
nnon_zero = nnon_zero.reset_index(drop=True)
nnon_zero
```

```python
#use this
for time in times[1:1000]:
    incides = nnon_zero["timestamp_sgt"][nnon_zero["timestamp_sgt"]==time].index.tolist()
    new_time = time.strftime("%Y-%m-%d %H")
    for index in incides:
        nnon_zero.set_value(index,"timestamp_sgt", new_time)
nnon_zero
```
